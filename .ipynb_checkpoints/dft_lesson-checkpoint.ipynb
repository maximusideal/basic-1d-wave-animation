## Discrete Fourier Transform

The Fourier transform and its inverse are given as
$$ \hat{\phi}(\omega) = \int_{-\infty}^{\infty} \phi(t)e^{-i\omega t}\, dt \qquad\text{and}\qquad \phi(t) = \frac{1}{2\pi}\int_{-\infty}^{\infty} \hat{\phi}(\omega)e^{i\omega t}\, d\omega. $$
When we compute these in Python, we have to discretize these, so we write
$$ \hat{\phi}(\omega) = \sum_{m=-M}^{M-1} \phi(m\Delta t)e^{-i\omega(m\Delta t)} \, \Delta t \qquad\text{and}\qquad \phi(t) = \frac{1}{2\pi}\sum_{n=-N}^{N-1} \hat{\phi}(n\Delta\omega)e^{i(n\Delta\omega) t}\, \Delta\omega. $$
Two key factors come into play: the sampling interval in the time domain $\Delta t$ and its counterpart in the frequency domain $\Delta\omega$. Most crucially, we want the inverse Fourier transform of the Fourier transform to give back the same function values at the sampled points. 

\begin{align}
\phi(m\Delta t) &= \frac{1}{2\pi}\sum_{n=-N}^{N-1} \hat{\phi}(n\Delta\omega)e^{i(n\Delta\omega)(m\Delta t)} \,\Delta\omega \\
&= \frac{1}{2\pi}\sum_{n=-N}^{N-1} \left( \sum_{m'=-M}^{M-1} \phi(m'\Delta t)e^{-i(n\Delta\omega)(m'\Delta t)} \,\Delta t \right) e^{i(n\Delta\omega)(m\Delta t)t} \Delta\omega \\
&= \frac{\Delta t\Delta\omega}{2\pi} \sum_{m'=-M}^{M-1} \phi(m'\Delta t) \sum_{n=-N}^{N-1} e^{i(n\Delta\omega)((m-m')\Delta t)} \\
&= \frac{\Delta t\Delta\omega}{2\pi} \sum_{m'=-M}^{M-1} \phi(m'\Delta t) \sum_{n=-N}^{N-1} e^{i(\Delta t\Delta\omega(m-m'))n} \\
&= \frac{\Delta t\Delta\omega}{2\pi} \sum_{m'=-M}^{M-1} \phi(m'\Delta t) \begin{cases}
    2N &\text{ if } m'=m \\
    \dfrac{2i\sin(\Delta t\Delta\omega(m-m')N)}{e^{i\Delta t\Delta\omega(m-m')} - 1} &\text{ otherwise.}
\end{cases}
\end{align}

To get back $\phi(m\Delta t)$ here, we will need to have $\Delta t\cdot\Delta\omega\cdot N = \pi$ or equivalently $N\Delta\omega = \pi/\Delta t$. 


We can interpret this result as follows. Consider the **Nyquist frequency**, defined as $f_{Ny} = 1 / (2\Delta t)$ or equivalently $\omega_{Ny} = \pi/\Delta t$ in angular terms. This is the highest frequency that can be accurately represented given a sampling interval $\Delta t$. Frequencies above the Nyquist frequency cannot be correctly sampled and will alias. Therefore, the range of frequencies we should consider for a discrete Fourier transform extends from $\omega_{\text{min}}=-\pi/\Delta t$ to $\omega_{\text{max}}=\pi/\Delta t$, ensuring we remain within bounds that prevent aliasing. Plugging $\omega_{\text{max}}=N\Delta\omega$ gives us the same relationship we found above. 


Observe also that if $m - m' = 2\pi k/(\Delta t\Delta\omega)$ where $k$ is a nonzero integer, then our inner sum in the second to last line is necessarily $2N$. In going from the second to last line to the last line, we implicitly assumed $2M-1 < 2\pi / \Delta t\Delta\omega$. Combined with our "Nyquist condition," we must have
$$ 2M-1 < \frac{2\pi}{\Delta t\Delta\omega} = \frac{2\pi}{\pi/N} = 2N, $$
or more simply,
$$ M < N+1. $$
If this condition were not in place, then our "recovered function" $\phi_{rec}$ would have had "improper repititions" of the form $\phi_{rec}(m\Delta t) = \phi_{rec}(m\Delta t + 2N\Delta t)$. This can be avoided by setting $M = N$. 


To summarize, our conditions are as follows:
1. $N\Delta\omega = \pi/\Delta t$ and
2. $M = N$.

The smaller $\Delta t$, the greater $\omega_{\text{max}}=\Delta\omega\cdot N$. The smaller $\Delta\omega$, the greater $t_{\text{max}} = \Delta t\cdot M$, as expected. Note we can relax condition 2 by taking $M \le N$ if we wished, although then we would might run into trouble recovering $\hat{\phi}$ when applying inverse FT followed by the FT on $\hat{\phi}$. 

### Initialization

```python
# import
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define function
def phi(t):
    sigma = 5
    return np.exp(-sigma * t**2)
```

### Example of Incorrect Sampling: len(t_array) < len(omega_array)


In this notebook, we will demonstrate how an incorrect choice of sampling leads to aliasing. The inverse DFT will not provide us the same values of our function unless we adhere to taking $N\Delta\omega = \pi/\Delta t$.

```python
# define arrays
t_array = np.linspace(-5, 5, 500, endpoint=False) 
dt = t_array[1]-t_array[0]

omega_array = np.linspace(-5, 5, 1500, endpoint=False) 
domega = omega_array[1]-omega_array[0]
```

```python
# sample phi(t)
phi_t = phi(t_array)

# take FT to get phi_hat(omega)
phi_hat_omega = np.array(
    [np.sum(phi_t * np.exp(-1j*omega*t_array) * dt) for omega in omega_array]
)

# take inverse FT to get phi_recover(t)
phi_recover = np.array(
    [np.sum(phi_hat_omega * np.exp(1j*omega_array*t) * domega) / (2*np.pi) for t in t_array]
)
```

```python
# visualize the original function
plt.figure(figsize=(10, 4))
plt.plot(t_array, np.real(phi_t))
plt.xlabel('t')
plt.ylabel('phi')
plt.grid(True)
plt.show()

# visualize the recovered function
plt.figure(figsize=(10, 4))
plt.plot(t_array, np.real(phi_recover))
plt.xlabel('t')
plt.ylabel('phi_recover')
plt.grid(True)
plt.show()
```

```python
# visualize the spectrum
plt.figure(figsize=(10, 4))
plt.plot(omega_array, np.real(phi_hat_omega), color='r')
plt.xlabel('omega')
plt.ylabel('phi_hat_omega')
plt.grid(True)
plt.show()
```

### Example of Incorrect Sampling: len(t_array) > len(omega_array)


Now $N\Delta\omega = \pi/\Delta t$ holds but we fail to adhere to $M = N$. 

```python
# define arrays
t_array = np.linspace(-5, 5, 500, endpoint=False) 
dt = t_array[1]-t_array[0]

omega_array = np.linspace(-np.pi/dt, np.pi/dt, 125, endpoint=False) # ideally set num of point to len(t_array)
domega = omega_array[1]-omega_array[0]
```

```python
# sample phi(t)
phi_t = phi(t_array)

# take FT to get phi_hat(omega)
phi_hat_omega = np.array(
    [np.sum(phi_t * np.exp(-1j*omega*t_array) * dt) for omega in omega_array]
)

# take inverse FT to get phi_recover(t)
phi_recover = np.array(
    [np.sum(phi_hat_omega * np.exp(1j*omega_array*t) * domega) / (2*np.pi) for t in t_array]
)
```

```python
# visualize the original function
plt.figure(figsize=(10, 4))
plt.plot(t_array, np.real(phi_t))
plt.xlabel('t')
plt.ylabel('phi')
plt.grid(True)
plt.show()

# visualize the recovered function
plt.figure(figsize=(10, 4))
plt.plot(t_array, np.real(phi_recover))
plt.xlabel('t')
plt.ylabel('phi_recover')
plt.grid(True)
plt.show()
```

```python
# visualize the spectrum
plt.figure(figsize=(10, 4))
plt.plot(omega_array, np.real(phi_hat_omega), color='r')
plt.xlabel('omega')
plt.ylabel('phi_hat_omega')
plt.grid(True)
plt.show()
```

### Example of Correct Sampling


Now we shall try to adhere to $M = N$ and $N\Delta\omega = \pi/\Delta t$. 

```python
# define arrays
t_array = np.linspace(-5, 5, 500, endpoint=False) 
dt = t_array[1]-t_array[0]

omega_array = np.linspace(-np.pi/dt, np.pi/dt, 500, endpoint=False) # ideally set num of point to len(t_array)
domega = omega_array[1]-omega_array[0]
```

```python
# sample phi(t)
phi_t = phi(t_array)

# take FT to get phi_hat(omega)
phi_hat_omega = np.array(
    [np.sum(phi_t * np.exp(-1j*omega*t_array) * dt) for omega in omega_array]
)

# take inverse FT to get phi_recover(t)
phi_recover = np.array(
    [np.sum(phi_hat_omega * np.exp(1j*omega_array*t) * domega) / (2*np.pi) for t in t_array]
)
```

```python
# visualize the original function
plt.figure(figsize=(10, 4))
plt.plot(t_array, np.real(phi_t))
plt.xlabel('t')
plt.ylabel('phi')
plt.grid(True)
plt.show()

# visualize the recovered function
plt.figure(figsize=(10, 4))
plt.plot(t_array, np.real(phi_recover))
plt.xlabel('t')
plt.ylabel('phi_recover')
plt.grid(True)
plt.show()
```

```python
# visualize the spectrum
plt.figure(figsize=(10, 4))
plt.plot(omega_array, np.real(phi_hat_omega), color='r')
plt.xlabel('omega')
plt.ylabel('phi_hat_omega')
plt.grid(True)
plt.show()
```
