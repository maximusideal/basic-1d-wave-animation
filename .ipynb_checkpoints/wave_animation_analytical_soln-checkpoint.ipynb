```python
# import
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
# constants
L = 1
c = 1
omega = 3 * np.pi
k = 1
totaltime = 5
```

```python
# define beta and gamma
beta = (1 / c) * np.sqrt((omega**2 + np.sqrt(omega**4 + omega**2 * k**2)) / 2)
gamma = -(np.sign(omega * k) / c) * np.sqrt((-omega**2 + np.sqrt(omega**4 + omega**2 * k**2)) / 2)

# define the function u(x, t)
def u(x, t):
    numerator = np.exp(gamma * (2*L - x)) * np.exp(1j * beta * x) - np.exp(1j * beta * (2*L - x)) * np.exp(gamma * x)
    denominator = np.exp(2 * gamma * L) - np.exp(2 * 1j * beta * L)
    return np.exp(1j * omega * t) * (numerator / denominator)
```

```python
# domain
x = np.linspace(0, L, 100)

# initial plot
fig, ax = plt.subplots()
line, = ax.plot(x, np.real(u(x, 0)), color='blue')
ax.set_xlim(0, 1)
ax.set_ylim(-2, 2)  # Adjust the y-axis limits if necessary
ax.set_title('Real part of u(x, t)')
ax.set_xlabel('x')
ax.set_ylabel('Real part of u(x, t)')

# update function for animation
def update(t):
    line.set_ydata(np.real(u(x, t)))  # Update the y data of the plot
    return line,

# create animation
ani = FuncAnimation(fig, update, frames=np.linspace(0, totaltime, 500), blit=True)

# save the animation (requires ffmpeg installed)
ani.save('u_x_t_animation.mp4', fps=30, extra_args=['-vcodec', 'libx264'])
plt.show()
```
