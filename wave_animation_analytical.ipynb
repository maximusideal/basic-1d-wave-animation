# Damped Wave Animation by Fourier Transforms


We want to compute a solution $u(x, t)$ satisfying the damped wave equation $\partial_{t}^{2}u + k\partial_{t}u = c^{2}\partial_{x}^{2}u$ with boundary conditions $u(x=0, t) = \phi(t)$ and $u(x=L, t) = 0$ and unspecified initial conditions. 

We take apart $\phi(t)$ by its Fourier transform:

$$ \phi(t) = \frac{1}{2\pi}\int_{-\infty}^{\infty} \hat{\phi}(\omega)e^{i\omega t}\, d\omega \qquad\text{and}\qquad \hat{\phi}(\omega) = \int_{-\infty}^{\infty} \phi(t)e^{-i\omega t}\, dt. $$

For each left boundary function $\phi_{\omega}(t) = e^{i\omega t}$, we find a solution

$$ u(\omega, x, t) = e^{i\omega t}\frac{ e^{\gamma(2L-x)}e^{i\beta x} - e^{i\beta(2L-x)}e^{\gamma x} }{e^{2\gamma L} - e^{2i\beta L}} $$

where

$$ \beta = \frac{1}{c}\sqrt{\frac{\omega^{2} + \sqrt{\omega^{4} + \omega^{2}k^{2}}}{2}} \quad\text{and}\quad \gamma = -\frac{\text{sign}(\omega k)}{c}\sqrt{\frac{-\omega^{2} + \sqrt{\omega^{4} + \omega^{2}k^{2}}}{2}}. $$

Given these solutions, we obtain the main solution by putting them together:

$$ u(x, t) = \frac{1}{2\pi}\int_{-\infty}^{\infty} u(\omega, x, t)\hat{\phi}(\omega)\, d\omega. $$


```python
# import
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
# constants
L = 3
c = 1
k = 0.2
nx = 30
t_total_play = 20
fps_play = 30

# define arrays
x_array_domain = np.linspace(0, L, nx)

t_array_play = np.arange(0, t_total_play, 1/fps_play)
t_array = np.linspace(-50, 50, 500) # limit for t integration
dt = t_array[1]-t_array[0]

omega_array = np.linspace(-np.pi/dt, np.pi/dt, 500) # limit for omega integration
domega = omega_array[1]-omega_array[0]
```

```python
# define boundary function
def phi(t):
    sigma = 5
    return np.exp(-sigma * (t - 1)**2)

# sample phi(t), take FT to get phi_hat(omega)
phi_t = phi(t_array)
phi_hat_omega = np.array(
    [np.trapz(phi_t * np.exp(-1j * omega * t_array), x=t_array) for omega in omega_array]
)
```

```python
# visualize the spectrum
plt.figure(figsize=(10, 4))
plt.plot(omega_array, np.real(phi_hat_omega))
plt.xlabel('omega')
plt.ylabel('phi_hat_omega')
plt.grid(True)
plt.show()
```

```python
# define integrand
# create 3D grids for omega, x, and t
omega_grid, x_grid, t_grid = np.meshgrid(omega_array, x_array_domain, t_array_play, indexing='ij')

# beta and gamma grids
beta_grid = (1 / c) * np.sqrt(
    (omega_grid**2 + np.sqrt(omega_grid**4 + omega_grid**2 * k**2)) / 2
)
gamma_grid = -(np.sign(omega_grid * k) / c) * np.sqrt(
    (-omega_grid**2 + np.sqrt(omega_grid**4 + omega_grid**2 * k**2)) / 2
)

# u_omega grid
num_grid = (np.exp(gamma_grid * (2*L - x_grid)) * np.exp(1j * beta_grid * x_grid) -
    np.exp(1j * beta_grid * (2*L - x_grid)) * np.exp(gamma_grid * x_grid))
denom_grid = (np.exp(2 * gamma_grid * L) - 
    np.exp(2 * 1j * beta_grid * L))

u_omega_x_t = np.exp(1j * omega_grid * t_grid) * num_grid / denom_grid
```

```python
# integrate back to get u(x, t)
# broadcast phi_hat_omega across dimensions of omega_grid, x_grid, t_grid
phi_hat_omega_grid = phi_hat_omega[:, np.newaxis, np.newaxis]

# get integrand grid
integrand_grid = u_omega_x_t * phi_hat_omega_grid / (2 * np.pi)

# perform integration over omega
u_x_t = np.trapz(integrand_grid, x=omega_array, axis=0)
```

```python
# plot u(0, t) as a check
fig0, ax0 = plt.subplots()
ax0.plot(t_array_play, np.real(u_x_t[0, :]))
ax0.set_xlabel('t')
ax0.set_ylabel('u(0, t)')
plt.show()
```

```python
# initial plot
fig, ax = plt.subplots()
initial_data = u_x_t[:, 0]
line, = ax.plot(x_array_domain, np.real(initial_data), color='blue')
ax.set_xlim(0, L)
ax.set_ylim(-2, 2)
ax.set_title(f'Re u(x, t) at t = 0.00, k = {k}, c = {c}')
ax.set_xlabel('x')
ax.set_ylabel('')

# update function for animation
def update(index):
    y_data = np.real(u_x_t[:, index])
    line.set_ydata(np.real(y_data))

    t_curr = t_array_play[index]
    prog = index/len(t_array_play)
    ax.set_title(f'Re u(x, t) at t = {t_curr:.2f}, k = {k}, c = {c}')
    
    print(f"\rprogress = {prog:.2f}           ", end = "")
    return line,

# create animation
ani = FuncAnimation(fig, update, frames=range(len(t_array_play)), blit=True)

# save the animation (requires ffmpeg installed)
ani.save('gaussianwave_animation1.mp4', fps=fps_play, extra_args=['-vcodec', 'libx264'])
plt.show()
```
